# Titles
In order to enter a title use a single pound symbol: # Title
## Subtitle
To enter a subtitle use two pound symbols: ## Subtitle
###### This is a subtitle? or something
To do this we use six pound symbols: ###### Subtitle or something...

# Formatting
## Emphasis
*Italics* or _Italics_ -> Make text italics by using * or _ as brackets. 
**Bold** or __Bold__ -> Make text bold by using ** or __ as brackets.
Combine bold and brackets: **bold and *bold and italics***

## Block Quotes
By using the arrow symbols we can create block quotes: 
> Whatever i type here will be in block quotes
>> This can span multiple lines or be intented further.

## Lists 
instead of using multiple asterisks we can use a single asterisks for a list
* Item one
* Item two
 * Item 2a
 * Item 2b 
   * Item 2bi

or we can also use a numbered list as we might expect

1. Item 1
2. Item 2
   * Item 2a

## Images
using the format: \!\[Alt Text\]\(url\) we can insert images:

![Python Logo](https://upload.wikimedia.org/wikipedia/commons/c/c3/Python-logo-notext.svg)

Links are automatically created so we can simply copy and paste them.

## Backslash Escapes
Since some of the commands use common symbols that we may want to include in our text we can simply use a backslach in front of them in order to escape using them as a command: \

## Line Divisions
Using three or more *** or --- or ___ we can create a line such as:
***

We can also experiment with other forms of line breaks such as 
One new line or:

Two new lines

___
---
